### 2.

**A scatter plot is a type of graph used to display the relationship between two continuous variables.**

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y,c='#fa4d8f')
plt.xlabel("This is my X data")
plt.ylabel("This is my Y data")
```

### 3.

**The subplot() function in matplotlib is used to create multiple plots within a single figure. This function allows you to create a grid of subplots, where you can specify the number of rows, columns, and the position of each subplot within the grid.**

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, a = plt.subplots(2,2)
fig.suptitle("Four Lines Of Plots")

a[0, 0].plot(x1,y1)
a[0, 0].set_title('Line 1')
                 
a[0, 1].plot(x2,y2)
a[0, 1].set_title('Line 2')
                 
a[1, 0].plot(x3,y3)
a[1, 0].set_title('Line 3')
                 
a[1, 1].plot(x4,y4)
a[1, 1].set_title('Line 4')
                 
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.85, wspace=0.4, hspace=0.4)
                 
plt.show()
```

### 4.

**A bar plot is a type of chart that displays categorical data with rectangular bars, where the height or length of each bar represents the value of the corresponding category. Bar plots are commonly used to compare and visualize the distribution or frequency of different categories or groups.**

**It is commonly used to show the frequency or proportion of different categories, or to compare the values of different groups within a category.**

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit, color='#fa4d8f')
plt.xlabel("This is Company Information")
plt.ylabel("This is Profit Information")
plt.grid()
```

```python
plt.barh(company,profit, color='#fa4d8f')
plt.xlabel("This is Profit Information")
plt.ylabel("This is Company Information")
plt.grid()
```

### 5.

**A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a set of numerical data through their quartiles. It displays the range, median, and quartiles of a dataset, and any potential outliers.**

**A typical box plot consists of a box that represents the interquartile range (IQR), with a line inside that indicates the median. Whiskers that extend from the box to indicate the range of the data, and any outliers or extreme values beyond the whiskers are displayed as individual points or dots.**

```python
import numpy as np
import matplotlib.pyplot as plt 

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

d = [box1,box2]
fig, a = plt.subplots()

a.boxplot(d)
a.set_title("Box Plots")
plt.show()
```
